```python
#Till line 14 its creating account and
class Bank:
    bankname= "Indian Bank"
    branch= "A23, Pune, India"

    def __init__(self, username, pan, address):
        self.username= username
        self.pan= pan
        self.address= address
        self.balance= 0.0 #as opening new account so balance will be zero
        self.acc_no= random.randint(1000000,9999999)
        self.history= []
        print(f"\nHello {self.username} your account has created succesfully!!")
        print(f"Account number: {self.acc_no}")

    def Deposit(self, amount):
        if amount <=0:
            print("Invalid amount. Deposit must be greater than zero")
            return # Added return to prevent further execution if amount is invalid
        self.balance +=amount
        self.history.append(f"Deposited {amount}")
        print(f"{amount} deposited successfully")

    def withdrawl(self, amount):
        if amount <=0:
            print("Invalid amount. withdrawl amount should be greater than 0")
            return
        if amount <= self.balance:
            self.balance -= amount
            self.history.append(f"Withdrawl {amount}")
            print(f"{amount} withdrawl successfully!!")
        else:
            print("Insufficient balance")

    def ministatement(self):
        print("\nMini Statement")
        print(f"Account Holder: {self.username}")
        print(f"Account number: {self.acc_no}")
        print(f"Address: {self.address}")
        print(f"PAN: {self.pan}")
        print("\nTransaction history:")
        if not self.history:
            print("No transactions yet")
        else:
            for transaction in self.history:
                print("-", transaction)
            print(f"\nCurrent Balance: {self.balance}")

        print(f"Current balance: {self.balance}")
#Account creation
username= input("Enter your name: ")
pan= input("Enter your PAN number: ")
address= input("Enter your address: ")

#Creation of object of class bank
user1= Bank(username, pan, address)

#main loop
while True:
    print("\nPlease choose an option") # Fixed escape sequence
    print("1.Deposit")
    print("2.Withdrawl")
    print("3.Mini Statement")
    print("4.Exit")

    options= int(input("Enter your option: "))
    if options==1: #
        amount= float(input("Enter deposit amount:"))
        user1.Deposit(amount)
    if options==2:
        amount= float(input("Enter withdrawl amount: "))
        user1.withdrawl(amount)
    elif options==3:
        user1.ministatement()

    elif options==4: # Changed 'option' to 'options'
      print("Thank you for banking with us!!")
      break
```
